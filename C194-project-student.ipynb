# Task 1 - Plot a line graph to show the difference between heart rate of smokers and non smokers

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('framingham.csv')
df
```

```python
nosmoke = df.loc[df['currentSmoker'] == 0]
nosmoke
```

```python
nosmoke_age = nosmoke.groupby('age')['heartRate'].mean().reset_index()
nosmoke_age
```

```python
smoke = df.loc[df['currentSmoker'] == 1]
smoke
```

```python
smoke_age = smoke.groupby('age')['heartRate'].mean().reset_index()
smoke_age
```

```python
nosmoke_label = nosmoke_age['age']
nosmoke_value = nosmoke_age['heartRate']

fig = plt.figure(figsize=(19,8))

plt.plot(nosmoke_label, nosmoke_value, label="Avg. Heart Rate of a non-smoker")

plt.legend()

smoke_label = smoke_age['age']
smoke_value = smoke_age['heartRate']

fig = plt.figure(figsize=(19,8))

plt.plot(smoke_label, smoke_value, label="Avg. Heart Rate of a smoker")

plt.xlabel("Age")
plt.xticks(rotation = "vertical")

plt.legend()
plt.show()
```

# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('framingham.csv')
df

chd = df.loc[df['TenYearCHD'] == 1]
chd
```

```python
age_group = chd.groupby('age')['TenYearCHD'].count().reset_index()
age_group
```

```python
fig = plt.figure(figsize=(19,8))

plt.scatter(age_group['age'], age_group['TenYearCHD'], label = "No. of CHD cases in different age groups", color = 'red', alpha=0.9, 
           s = age_group['TenYearCHD']*5)

plt.xlabel("age")
plt.ylabel("Risk")
plt.xticks(rotation='vertical')
plt.legend()
plt.show()
```
